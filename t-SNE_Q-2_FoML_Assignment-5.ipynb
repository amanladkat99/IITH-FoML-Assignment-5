```python
from sklearn.datasets import load_digits
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
digits = load_digits()
```

```python
X = digits.data[:]
X.shape
```

```python
Y = digits.target[:]
Y.shape
```

# Perplexity = 30 and Number of iterations = 1000

```python
tsne = TSNE(n_components=2, perplexity=30, n_iter=1000, random_state=None)
X_2d = tsne.fit_transform(X)
digits_ids = range(len(digits.target_names))

plt.figure(figsize=(10, 8))
colors = 'aqua', 'blue', 'gray', 'gold', 'green', 'fuchsia', 'maroon', 'purple', 'red', 'orange'
for i, c, label in zip(digits_ids, colors, digits.target_names):
    plt.scatter(X_2d[Y == i, 0], X_2d[Y == i, 1], c=c, label=label)
plt.title("Number of Iterations = 1000")
plt.legend()
plt.show()
```

# Perplexity = 30 and Number of iterations = 250, 2000 and 5000
                                   (Should be at least 250)

```python
tsne = TSNE(n_components=2, perplexity=30, n_iter=250, random_state=None)
X_2d = tsne.fit_transform(X)
digits_ids = range(len(digits.target_names))

plt.figure(figsize=(10, 8))
colors = 'aqua', 'blue', 'gray', 'gold', 'green', 'fuchsia', 'maroon', 'purple', 'red', 'orange'
for i, c, label in zip(digits_ids, colors, digits.target_names):
    plt.scatter(X_2d[Y == i, 0], X_2d[Y == i, 1], c=c, label=label)
plt.title("Number of Iterations = 250")
plt.legend()
plt.show()
```

# (a)

Answer : As the number of iterations increases from 250 to 5000, in t-SNE, the scatterplot quality improves as seen above with different values. With lower iterations, massive overlaps and random fragmentation have been observed in the plot.

# (b)

Answer : Basically, t-SNE focuses on retaining the structure of neighbour points. t-SNE employs gradient descent to minimize the  divergence of two distributions. It is a algorithm with random probability, which concerned only about retaining the variance of neighbour points. As, this is a non-convex optimization problem, hence we can get different results with the same
hyperparameters for some datasets.
